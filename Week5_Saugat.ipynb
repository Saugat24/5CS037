```python
#To-Do Task 4: Building a Cost Function (MSE)

def cost_function(X, Y, W):
    m = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * m)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
#To-Do Task 5: Designing a Test Case for Cost Function

X_test_case = np.array([[1, 2],
                        [3, 4],
                        [5, 6]])

Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])

cost = cost_function(X_test_case, Y_test_case, W_test_case)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")

print("Cost:", cost)
```

```python
#To-Do Task 6: Implementing Gradient Descent

def gradient_descent(X, Y, W, alpha, iterations):
    m = len(Y)
    cost_history = []

    for _ in range(iterations):
        # Prediction
        Y_pred = np.dot(X, W)

        # Loss
        loss = Y_pred - Y

        # Gradient
        dw = (1 / m) * np.dot(X.T, loss)

        # Update weights
        W = W - alpha * dw

        # Cost
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python
#To Do Task 7: Gradient Descent Test Case

np.random.seed(0)
X_rand = np.random.rand(100, 3)
Y_rand = np.random.rand(100)
W_rand = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(
    X_rand, Y_rand, W_rand, alpha, iterations
)

print("Final Parameters:", final_params)
print("Final Cost:", cost_history[-1])
```

```python
#To Do Task 8: Implementation of RMSE

def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

```python
#To Do Task 8: Implementation of R^2

def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
#To Do Task 9: Main Function to Integrate All Steps

def main():
    # Load data
    data = pd.read_csv("/content/drive/MyDrive/Dataset/student.csv")

    # Split the data into features (X) and target (Y)
    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    # Train-test split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Initialize parameters
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Gradient Descent
    W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    # Predictions on the test set
    Y_pred = np.dot(X_test, W_optimal)

    #Evaluate the model using RMSE and R-Squared
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Evaluation
    print("Final Weights:", W_optimal)
    print("Cost History (First 5 iterations):", cost_history[:5])
    print("RMSE:", model_rmse)
    print("R² Score:", model_r2)

if __name__ == "__main__":
    main()
```
