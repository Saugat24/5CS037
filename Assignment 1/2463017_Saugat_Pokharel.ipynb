```python
#Task6: Adding HDI Category Column:

def classify_hdi(hdi):
    if hdi < 0.550:
        return 'Low'
    elif 0.550 <= hdi <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi <= 0.799:
        return 'High'
    else: # hdi >= 0.800
        return 'Very High'

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)
hdi_2022_df['HDI Category'].value_counts()
```

```python
#B. Generate Visualizations:
# Bar Chart: Average HDI by Region (2020–2022)
   #Group the dataset by Region and Year.
   #Compute the mean HDI for each region-year pair.
   #Plot a bar chart comparing average HDI across regions for each year.
   #Label axes clearly and include a descriptive title.

region_avg_hdi = (
    hdi_1b_df
    .groupby(['region', 'year'])['hdi']
    .mean()
    .reset_index()
)

region_avg_hdi
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=region_avg_hdi,
    x='region',
    y='hdi',
    hue='year'
)

plt.xlabel("Region")
plt.ylabel("Average HDI")
plt.title("Average HDI by Region (2020–2022)")
plt.legend(title="Year")
plt.xticks(rotation=30)
plt.show()

print("The bar chart shows clear regional disparities in HDI levels between 2020 and 2022.")
print("Europe and North America consistently record higher average HDI values, while South Asia and Africa remain lower.")
print("A slight decline or stagnation is observed around 2020–2021, likely due to the COVID-19 pandemic, followed by modest recovery in 2022.")
```

```python
#3 Outlier Detection:
#Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
#Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.
#Discuss why the identified countries stand out as outliers.
```

```python
#Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
# IQR for HDI
Q1_hdi = south_asia_latest['hdi'].quantile(0.25)
Q3_hdi = south_asia_latest['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# IQR for GNI per Capita
Q1_gni = south_asia_latest['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_latest['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# Identify outliers
outliers = south_asia_latest[
    (south_asia_latest['hdi'] < lower_hdi) |
    (south_asia_latest['hdi'] > upper_hdi) |
    (south_asia_latest['gross_inc_percap'] < lower_gni) |
    (south_asia_latest['gross_inc_percap'] > upper_gni)
]

outliers[['country', 'hdi', 'gross_inc_percap']]
```

```python
#Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.
#Using the 1.5 × IQR rule, no countries in South Asia were identified as outliers in either HDI or GNI per Capita.
#This suggests that development levels within the region are relatively homogeneous, with no extreme deviations during the selected year.

plt.figure(figsize=(6,5))

# Normal points
plt.scatter(
    south_asia_latest['gross_inc_percap'],
    south_asia_latest['hdi'],
    label="Normal Countries"
)

# Outliers
plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    color='red',
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (South Asia) with Outliers")
plt.legend()
plt.show()
```

```python
#Discuss why the identified countries stand out as outliers.

print("The scatter plot reveals countries that deviate noticeably from regional trends in income and human development.")
print("Nations with exceptionally high GNI per capita but only moderate HDI emerge as positive income outliers,")
print("while countries with both low income and low HDI appear as negative outliers.")
print("Such deviations can be attributed to factors like income inequality,")
print("variations in government spending on health and education, political instability,")
print("and underlying structural features of the economy.")
```

```python
#6 Correlation Analysis:

#For each region, compute correlations of HDI with:
  #– Gender Development Index
  #– Life Expectancy Index
#Create scatter plots with trendlines for each correlation.
#Interpret the strength and direction of these relationships.
```

```python
def correlation_analysis(df, region):
    corr_gender = df['hdi'].corr(df['gender_development'])
    corr_life   = df['hdi'].corr(df['life_expectancy'])

    print(f"{region} Correlations")
    print(f"HDI vs Gender Development: {corr_gender:.3f}")
    print(f"HDI vs Life Expectancy   : {corr_life:.3f}")

    # Scatter plots
    for col in ['gender_development', 'life_expectancy']:
        plt.figure()
        plt.scatter(df[col], df['hdi'])
        z = np.polyfit(df[col], df['hdi'], 1)
        p = np.poly1d(z)
        plt.plot(df[col], p(df[col]))
        plt.xlabel(col)
        plt.ylabel("HDI")
        plt.title(f"{region}: HDI vs {col}")
        plt.show()

correlation_analysis(south_asia_2020_2022, "South Asia")
correlation_analysis(middle_east_2020_2022, "Middle East")
```

```python
#7 Outlier Detection:

#Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
#the 1.5 × IQR rule.
#Create scatter plots highlighting outliers in a different color.
#Discuss the significance of these outliers.
```

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] < Q1 - 1.5*IQR) | (df[column] > Q3 + 1.5*IQR)]

for region, data in [("South Asia", south_asia_2020_2022),
                     ("Middle East", middle_east_2020_2022)]:

    out_hdi = detect_outliers(data, 'hdi')
    out_gni = detect_outliers(data, 'gross_inc_percap')

    plt.figure()
    plt.scatter(data['gross_inc_percap'], data['hdi'])
    plt.scatter(out_gni['gross_inc_percap'], out_gni['hdi'])
    plt.xlabel("GNI per Capita Index")
    plt.ylabel("HDI")
    plt.title(f"{region}: Outliers in HDI & GNI")
    plt.show()
```

```python
print("HDI and GNI per capita outliers reveal countries that deviate from regional patterns.\n")
print("- Some nations have high income but only moderate HDI, suggesting that economic wealth is not always effectively")
print("  translated into better health, education, and overall well-being.\n")
print("- Conversely, countries with lower income but relatively high HDI demonstrate that strong social policies")
print("  and investments in human capital can lead to impressive development outcomes.\n")
print("Recognizing these outliers helps guide development strategies, understand regional inequalities,")
print("and draw lessons from both exceptional performers and underperformers.")
```
